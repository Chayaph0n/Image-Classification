```python
from PIL import Image
import matplotlib.pyplot as plt
import os 
from os.path import isfile, join
from os import listdir
from tqdm import tqdm
```

```python
# Define the transformation functions
def save_and_display_image(image, filename, type):
    directory = '../content/image_test/'
    
    if type == '2' :
        save_directory = os.path.join(directory, '2')
        image.save(os.path.join(save_directory, filename))
        print('save in folder 2 : ' + str(os.path.join(save_directory, filename)))
    elif type == '3' :
        save_directory = os.path.join(directory, '3')
        image.save(os.path.join(save_directory, filename))
        print('save in folder 3 : ' + str(os.path.join(save_directory, filename)))
    elif type == '4' :
        save_directory = os.path.join(directory, '4')
        image.save(os.path.join(save_directory, filename))
        print('save in folder 4 : ' + str(os.path.join(save_directory, filename)))
    elif type == '5' :
        save_directory = os.path.join(directory, '5')
        image.save(os.path.join(save_directory, filename))
        print('save in folder 5 : ' + str(os.path.join(save_directory, filename)))

    # plt.imshow(image)
    # plt.title(filename)
    # plt.show()
```

```python
trainpath = '../content/Image_wo1&6/train/'
trainImg = [trainpath+f for f in listdir(trainpath)]

for imagePath in trainImg :
    for item in tqdm(listdir(imagePath)):
            file = join(imagePath, item)
            
            # Extract the filename without extension 'filename' + '.jpg'
            filename = os.path.splitext(os.path.basename(file))[0]

            # Extract the rotation and flip information
            serial = filename.split('_')[0]
            value_shear = filename.split('_')[1]
            type = filename.split('_')[2]

            # Open the input image
            input_image = Image.open(file)

            # Rotate the image by 90 degrees and save
            rotated_90 = input_image.rotate(90)
            new_filename = f"{serial}_rotated_90_{value_shear}_{type}.jpg"
            save_and_display_image(rotated_90, new_filename, type)

            # Rotate the image by 180 degrees and save
            rotated_180 = input_image.rotate(180)
            new_filename = f"{serial}_rotated_180_{value_shear}_{type}.jpg"
            save_and_display_image(rotated_180, new_filename, type)

            # Rotate the image by 270 degrees and save
            rotated_270 = input_image.rotate(270)
            new_filename = f"{serial}_rotated_270_{value_shear}_{type}.jpg"
            save_and_display_image(rotated_270, new_filename, type)

            # Flip the image horizontally (mirror image) and save
            flipped_x = input_image.transpose(Image.FLIP_LEFT_RIGHT)
            new_filename = f"{serial}_flipx_{value_shear}_{type}.jpg"
            save_and_display_image(flipped_x, new_filename, type)

            # Flip the image vertically (upside down) and save
            flipped_y = input_image.transpose(Image.FLIP_TOP_BOTTOM)
            new_filename = f"{serial}_flipy_{value_shear}_{type}.jpg"
            save_and_display_image(flipped_y, new_filename, type)

            # Close the input image
            input_image.close()
```
